# Examining image datasets in fast.ai
Walkthrough of how to examine image datasets using the facilities in fast.ai.


```python
# imports for notebook boilerplate
!pip install -Uqq fastbook
import fastbook
from fastbook import *
from fastai.vision.all import *
```

```python
# set up the notebook for fast.ai
fastbook.setup_book()
```

```python
# ingest the curated image dataset FLOWERS
path = untar_data(URLs.FLOWERS)
```

```python
# examine the directory structure of the dataset
path.ls()
```

```python
# ingest the train dataset list into a Pandas dataframe

df_valid = pd.read_csv(path/'valid.txt', header=None)
df_valid.head()
```

```python
# examine the jpg subdirectory
(path/'jpg').ls()
```

```python
# take a look at one of the images
img_files = get_image_files(path)
img = PILImage.create(img_files[100])
img
```

```python
# display a thumbnail of the image
img.to_thumb(180)
```
